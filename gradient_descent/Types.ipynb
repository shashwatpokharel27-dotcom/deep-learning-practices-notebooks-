Batch Gradient Descent -> All Data → Compute Gradient → Update Weights (normal hamile gareko jasto)

✅ Stable updates

✅ Accurate gradient

❌ Very slow for large datasets

❌ Requires lots of memory

Stochastic Gradient Descent (SGD) -> "Stochastic" means random.

Instead of using all data:

Take 1 sample.
Update weights.
Take next sample.
Update again.

Mini-Batch Gradient Descent -> Instead of:

All samples (Batch GD)
One sample (SGD)

Use a small group:

32 samples
64 samples
128 samples
256 samples

| Method     | Speed         | Stability    |
| ---------- | ------------- | ------------ |
| Batch GD   | Slow          | Very Stable  |
| SGD        | Fast Updates  | Noisy        |
| Mini-Batch | Fast + Stable | Best Balance |
