```python
X = df.drop(60, axis=1)
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R --> 1 and M --> 0
```

```python
y.value_counts()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.head()
```

```python
X_train.shape
```

# Using Deep Learning Model

## Model without Dropout Layer

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

The 156 samples are split into groups of 8:


Epoch 1
 в”њв”Ђ Batch 1 (8 samples)
 в”њв”Ђ Batch 2 (8 samples)
 в”њв”Ђ ...
 в””в”Ђ Batch 20 (last batch has 4 samples)

Weights updated 20 times

For each batch:

Forward pass (make predictions),
Calculate loss,
Backpropagation (compute gradients),
Optimizer updates weights.

Then move to the next batch.

100 epochs Г— 20 batches
= 2000 weight updates

Called Batch Gradient Descent

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
print(y_pred[:10])
```

```python
# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
print(classification_report(y_test, y_pred))
```

### Model with Dropout Layer

```python
modeld=keras.Sequential([keras.layers.Dense(60,input_dim=60,activation='relu'),
                         keras.layers.Dropout(0.5),
                         keras.layers.Dense(30, activation='relu'),
                         keras.layers.Dropout(0.5),
                         keras.layers.Dense(30, activation='relu'),
                         keras.layers.Dropout(0.5),
                         keras.layers.Dense(30, activation='relu')])
```

```python
modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
modeld.evaluate(X_test, y_test)
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```
