```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}') 
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
```

```python
df2.head(2)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df2[cols_to_scale]=scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

train test split

```python
X = df2.drop('Churn',axis='columns')
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

**Build a model (ANN) in tensorflow/keras**

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([keras.layers.Dense(26,input_shape=(26,),activation='relu'),
                        keras.layers.Dense(15,activation='relu'),
                        keras.layers.Dense(1,activation='sigmoid')])
```

Dense(26, input_shape=(26,)) -> Each of the 26 neurons has:
26 weights

ReLU Formula

f(x)=max(0,x)

The sigmoid function converts any number into a value between 0 and 1.

For binary classification ,use sigmoid.For binary classification, binary cross-entropy is the standard choice

The optimizer updates the model's weights to reduce the loss.It decides how big a step to take and how to use the gradient information.

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
yp = model.predict(X_test)
```

```python
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
```

```python
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
