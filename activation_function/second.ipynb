Vanishing gradient is a problem during training of neural networks where the gradients (updates to weights) become extremely small—almost zero.

it happens: Because:

sigmoid and tanh derivatives are less than 1
When multiplied repeatedly across many layers →
the gradient shrinks exponentially

0.5 × 0.5 × 0.5 × 0.5 × ... → very close to 0

which causes, they learn very slowly or not at all

Leaky ReLU is a small improvement over standard ReLU designed to fix the “dying ReLU” problem.

Instead of making all negative inputs zero (like ReLU), it allows a small slope for negative values:

In standard ReLU:
Negative inputs → output = 0 → gradient = 0 → neuron stops learning
In Leaky ReLU:
Gradient is small but non-zero → neuron can still learn
